```python
colunas_desejadas = [
    'NU_INSCRICAO',
    'NU_ANO',
    'TP_FAIXA_ETARIA',
    'TP_SEXO',
    'TP_ESTADO_CIVIL',
    'TP_COR_RACA',
    'TP_NACIONALIDADE',
    'TP_ESCOLA',
    'TP_ENSINO',
    'IN_TREINEIRO',
    'NO_MUNICIPIO_PROVA',
    'SG_UF_PROVA',
    'TP_DEPENDENCIA_ADM_ESC',
    'TP_LOCALIZACAO_ESC'
]

df_limpo = df_limpo[colunas_desejadas]

df_limpo.head()
```

```python
#dicionario mapeamento faixa etaria

faixa_etaria_map ={
     1: "Menor de 17 anos",
    2: "17 anos",
    3: "18 anos",
    4: "19 anos",
    5: "20 anos",
    6: "21 anos",
    7: "22 anos",
    8: "23 anos",
    9: "24 anos",
    10: "25 anos",
    11: "Entre 26 e 30 anos",
    12: "Entre 31 e 35 anos",
    13: "Entre 36 e 40 anos",
    14: "Entre 41 e 45 anos",
    15: "Entre 46 e 50 anos",
    16: "Entre 51 e 55 anos",
    17: "Entre 56 e 60 anos",
    18: "Entre 61 e 65 anos",
    19: "Entre 66 e 70 anos",
    20: "Maior de 70 anos"
}


# substituir categoria pela descrição
df_limpo ['TP_FAIXA_ETARIA'] = df['TP_FAIXA_ETARIA'].map(faixa_etaria_map)

#exibir mapeamento
df_limpo [['TP_FAIXA_ETARIA']].head()
```

```python
#mapeamento ddo dicionario de estado civil

estado_civil = {
    0:"Não informado",
    1:"Solteiro(a)",
    2:"Casado(a)/Mora com companheiro(a)",
    3:"Divorciado(a)/Desquitado(a)/Separado(a)",
    4:"Viúvo(a)"
}

df_limpo ['TP_ESTADO_CIVIL'] = df['TP_ESTADO_CIVIL'].map(estado_civil)
```

```python
#tipo cor/ raça

mapeamento_cor = {
    0: "Não declarado",
    1: "Branca",
    2: "Preta",
    3: "Parda",
    4: "Amarela",
    5: "Indígena",
    6: "Não dispõe da informação"

}

df_limpo['TP_COR_RACA'] = df['TP_COR_RACA'].map(mapeamento_cor)
df_limpo['TP_COR_RACA'].value_counts()
```

```python
df_limpo.head()
```

```python
#mapeamento nacionalidade

nacionalidade = {
    0: "Não informado",
    1: "Brasileiro(a)",
    2: "Brasileiro(a) Naturalizado(a)",
    3: "Estrangeiro(a)",
    4: "Brasileiro(a) Nato(a), nascido(a) no exterior"
}


df_limpo['TP_NACIONALIDADE'] = df_limpo['TP_NACIONALIDADE'].map(nacionalidade)

df_limpo['TP_NACIONALIDADE'].value_counts()
```

```python
# mapeamento tipo escola

tipo_escola = {
    1: 'Não Respondeu',
    2: 'Pública',
    3: 'Privada'
}

df_limpo['TP_ESCOLA'] = df_limpo['TP_ESCOLA'].map(tipo_escola)

df_limpo['TP_ESCOLA'].value_counts()
```

```python
# mapeamento de distribuição escolar

mapeamento_dependencia = {
    1: 'Federal',
    2: 'Estadual',
    3: 'Municipal',
    4: 'Privada'
}

df_limpo['TP_DEPENDENCIA_ADM_ESC'] = df_limpo['TP_DEPENDENCIA_ADM_ESC'].map(mapeamento_dependencia)

df_limpo['TP_DEPENDENCIA_ADM_ESC'].value_counts()
```

```python
df_limpo.head()
```

```python
# mapeamento Tipo de instituição que concluiu ou concluirá o Ensino Médio

tipo_ensino = {
    1:"Ensino Regular",
    2:"Educação Especial - Modalidade Substitutiva"
}

df_limpo['TP_ENSINO'] = df['TP_ENSINO'].map(tipo_ensino)

df['TP_ENSINO'].value_counts()
```

```python
# mapeamento treineiro

mapeamento_treneiro = {
    1:"Sim",
    0:"Não"
}

df_limpo['IN_TREINEIRO'] = df['IN_TREINEIRO'].map(mapeamento_treneiro)

df['IN_TREINEIRO'].value_counts()
```

```python
#distribuição de cada coluna

print(df_limpo['TP_ENSINO'].value_counts())
print(df_limpo['IN_TREINEIRO'].value_counts())
```

```python
# mapeando localização escola

localizacao_escola = {
    "1.0": "Urbana",
    "2.0": "Rural"
}

df['TP_LOCALIZACAO_ESC'] = (
     df['TP_LOCALIZACAO_ESC']
     .map({1.0:"Urbana", 2.0:"Rural"})
     .astype('string')
)
df['TP_LOCALIZACAO_ESC'].value_counts()

```
